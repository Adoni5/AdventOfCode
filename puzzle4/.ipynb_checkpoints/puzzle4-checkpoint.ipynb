```python
%%timeit
import numpy as np
import pandas as pd
import re
pd.options.mode.chained_assignment = None
with open("puzzle4_input.txt") as file:
    
    lines = file.read()
    
    parsed = [re.findall(r"\[([\d]+)-([\d]+)-([\d]+)\s(\d{2}:\d{2})\]\s(?:Guard\s#(\d+)\s)?(?:.*(begin|asleep|wakes))", lines)][0]
    df = pd.DataFrame(parsed)
    df.columns=["year", "month", "day", "minute", "guard", "action"]
    def filler(col):
        return np.where(col=="", np.NaN, col)

    df["guard"] = filler(df["guard"].values)

    df = df.sort_values(["year", "month", "day", "minute"])
    df["guard"] = df["guard"].fillna(method="ffill")
    df = pd.merge(df, df["minute"].str.split(":", expand=True), left_index=True, right_index=True)
    test_df = df[df["action"].ne("begin")]
    test_df["minute"] = test_df[1].astype("int64")
    test_df.set_index("guard", inplace=True)
    sleep_series = test_df[test_df["action"].eq("asleep")]["minute"]
    woke_series = test_df[test_df["action"].eq("wakes")]["minute"]

    time_asleep_series = woke_df["minute"] - sleep_df["minute"]
    time_asleep_series.name = "time_asleep"



    # Part 1
    sleepiest_guard = time_asleep_series.groupby("guard").sum().sort_values(ascending=False).index[0]
    sleepiest_guard_rows = time_asleep_series[time_asleep_series.index == sleepiest_guard]

    time_asleep_df = pd.concat([sleep_series[sleep_series.index==sleepiest_guard], sleepiest_guard_rows], axis=1)

    time_asleep_df["end_time"] = time_asleep_df["minute"] + time_asleep_df["time_asleep"]

    minutes_asleep_list = [np.arange(x.minute, x.end_time, 1) for x in time_asleep_df.itertuples()]
    minutes_asleep_list = np.concatenate(minutes_asleep_list).ravel()
    sleepiest_minute = np.argmax(np.bincount(minutes_asleep_list))
    print(f"combined sleepiest minute with sleepiest guard: {sleepiest_minute * int(sleepiest_guard)}")



```

```python
total_time_asleep_df = pd.concat([sleep_series, time_asleep_series], axis=1)
total_time_asleep_df["end_time"] = total_time_asleep_df["minute"] + total_time_asleep_df["time_asleep"]
```

```python
gb = total_time_asleep_df.groupby(level="guard")
a = gb.apply(per_guard_time)
```

```python
def per_guard_time(group_df):
    minutes_asleep_list = [np.arange(x.minute, x.end_time, 1) for x in group_df.itertuples()]
    minutes_asleep_list = np.concatenate(minutes_asleep_list).ravel()
    unique, counts = np.unique(minutes_asleep_list, return_counts=True)
    all_mins_count = dict(zip(unique, counts))
    sleepiest_minute = np.argmax(np.bincount(minutes_asleep_list))
    return {sleepiest_minute: all_mins_count[sleepiest_minute]}

```
